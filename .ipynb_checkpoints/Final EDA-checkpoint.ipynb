```python
# converting date to datetime

weather['Date'] = pd.to_datetime(weather['Date'])
```

```python
# setting date to index

weather = weather.set_index('Date')
```

```python
# station 2 sucks at recording things

weather.head()
```

```python
# here we will replace 'M' with nan values (as it is missing) and change 'T' to a small decimal
# as it was recorded as 'trace'

weather.replace('M', weather.replace(['M'], [None]), inplace = True)
weather.replace('T', weather.replace(['T'], 0.005), inplace = True)
```

```python
weather.isnull().sum()
```

```python
# deleting the columns as discussed

weather.drop(columns = ['Water1', 'Depth', 'SnowFall', 'Depart', 'CodeSum'], inplace = True)
```

```python
# making the columns with null values easier to read

null_columns = weather.columns[weather.isnull().any()]
weather[null_columns].isnull().sum()
```
